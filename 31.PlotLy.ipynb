```python
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot
import numpy as np
from random import uniform
```

```python
init_notebook_mode(connected=True)
```

##  Error Bars Animation

```python


x_theo = np.linspace(-4, 4, 100)
y_theo = np.sinc(x_theo)
x = [-3.8, -3.03, -1.91, -1.46, -0.89, -0.24, -0.0, 0.41, 0.89, 1.01, 1.91, 2.28, 2.79, 3.56]
y = [-0.02, 0.04, -0.01, -0.27, 0.36, 0.75, 1.03, 0.65, 0.28, 0.02, -0.11, 0.16, 0.04, -0.15]


```

```python
data_frames = []
data_y = []
for i in range(50):
    
    data_y.append([e+uniform(-0.1,0.1) for e in y ])
    data_frames.append((dict(x=x_theo,y=y_theo),
                        dict(x=x,
                            y=data_y[i],
                            mode='markers',
                            error_y=dict(
                                        type='constant',
                                        value=0.1,
                                        color='#85144B',
                                        thickness=1.5,
                                        width=3,
                                        opacity=1
                            ),
                            error_x=dict(
                                        type='constant',
                                        value=0.2,
                                        color='#85144B',
                                        thickness=1.5,
                                        width=3,
                                        opacity=1
    ),
                            marker=dict(
        color='#85144B',
        size=8)
                            )))

```

```python
import plotly.plotly as py
figure = {'data': [dict(x=x_theo,y=y_theo), dict(x=x,y=y)],
          'layout': {'xaxis': {'range': [-4, 4], 'autorange': False},
                     'yaxis': {'range': [-1.2, 1.2], 'autorange': False},
                     'title': 'Moving Error Bars',
                     'updatemenus': [{'type': 'buttons',
                                      'buttons': [{'label': 'Play',
                                                   'method': 'animate',
                                                   'args': [None]}]}]
                    }}



figure['frames'] = [dict(data=[f1,f2]) for f1,f2 in data_frames]

figure['frames']

iplot(figure)
```
