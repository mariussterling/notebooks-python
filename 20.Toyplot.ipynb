```python
import toyplot
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
y = 1-np.exp(-x)
toyplot.plot(x,y, width=400);
```

```python
import toyplot.locator

canvas = toyplot.Canvas(width=600, height=300)
axes = canvas.cartesian()
axes.label.text = "Condensator charge"
axes.x.label.text = "t"
axes.y.label.text = "Q(t)"
axes.x.ticks.show = True
mark = axes.plot(x, y)
```

```python
toyplot.fill(x, y,  width=600, height=200);
```

```python
edges = toyplot.generate.barabasi_albert_graph(n=10)
layout = toyplot.layout.FruchtermanReingold()
```

Of course, you can assign a $[0, N)$ colormap to the vertices based on their index, or some other variable:

```python
colormap = toyplot.color.LinearMap(toyplot.color.Palette(["white", "yellow", "red"]))
vstyle = {"stroke":toyplot.color.near_black}

toyplot.graph(edges, layout=layout, vcolor=colormap, vsize=30, vstyle=vstyle, width=500);
```

Edges can also be rendered with per-edge *head*, *middle*, and *tail* :ref:`markers <markers>`.  For example, if you wish to show the directionality of the edges in a graph, it is customary to add an arrow at the end of each edge:

```python
toyplot.graph(
    edges,
    layout=layout,
    ecolor="black",
    tmarker=">",
    vcolor=colormap,
    vsize=30,
    vstyle=vstyle,
    width=500,
);
```
